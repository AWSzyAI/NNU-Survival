```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_node("时子延")
```

```python
G.nodes()
```

```python
G.add_nodes_from([
    ("竹山路",{"color":"red"}),
    ("龙眠大道",{"color":"blue"}),
])
```

```python
G.nodes()
```

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
G.add_node(H)
G.nodes()
```

```python
G.clear()
G.nodes()
```

```python
G = nx.petersen_graph()
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```

```python
nx.draw(G)
plt.savefig("simple_path.png") # save as png
```
